```python
import pandas as pd
import keras
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras import optimizers
from tensorflow.keras import Sequential
```

```python
train = pd.read_csv("data/party_A_x_train.csv", delimiter=r",")
trainer = ["x0", "x1", "x2", "x3", "x4", "x5", "x6", "x7", "x8", "x9", "x10", "x11"]
train.head()
```

```python
model = Sequential()
model.add(Dense(units=1, activation='relu'))
model.compile(loss='mean_squared_error', optimizer='adam')
model.fit(train[trainer], train['y'],epochs=100, verbose=0)
```

```python
pred = model.predict(train[trainer])
```
