# Helper functions

```python
def format_FL_output(df):
    df['cycle'] = df['id'] % 1000
    df['unit num'] = (df['id'] - df['cycle'])/1000
    df['unit num'] = df['unit num'].astype(int)
    df.rename(columns={'predict_result': 'GBDT (FL)'}, inplace=True)
```

# Formatting federated output

```python
format_FL_output(df_A)
format_FL_output(df_B)
format_FL_output(df_C)
```

```python
df_master_FL = df_A.copy()
df_master_FL=df_master_FL.append(df_B)
df_master_FL=df_master_FL.append(df_C)
```

```python
# drop rows that are train
df_train_FL = df_master_FL[df_master_FL.type == "train"]
df_test_FL = df_master_FL[df_master_FL.type == "validate"]
```

# Plotting output

```python
# plot line graph
unit = [1, 2, 3, 4, 65, 66, 67, 68, 89, 90 , 91, 92]
fig, axs = plt.subplots(int(len(unit)/4), 4)  # row, col

i = 0
for ax in axs.flatten():
    ax.set_title("Engine " + str(unit[i]))
    
    # plot true RUL
    ax.plot(df_train_FL.loc[df_train_FL['unit num'] == unit[i], 'cycle'],  # x-axis
            df_train_FL.loc[df_train_FL['unit num'] == unit[i], 'GBDT (FL)'],      # y-axis      
            linewidth = 0.75, color="red", label='FL Predicted')
    
    # plot predict RUL
    ax.plot(df_train_FL.loc[df_train_FL['unit num'] == unit[i], 'cycle'],
            df_train_FL.loc[df_train_FL['unit num'] == unit[i], 'label'],
            linewidth = 0.75, color="skyblue", label='True RUL')
    i += 1
plt.setp(axs[-1, :], xlabel='Cycles')
plt.setp(axs[:, 0], ylabel='Normalised sensor readings')
plt.legend(prop={'size': 13})
fig.set_size_inches(15, 10)  # width x height
plt.show()

```
